# 1 Preparing the coding environment - importing libraries.

* Our plan is to use visualisations to answer certain questions about our data.

* We will start by importing the pandas and matplotlib libraries.
* We will read in the revent_grads.csv file

```python
#importing Pandas and Matplotlib Libraries
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')

print(recent_grads.iloc[0])
print('\n')
```

# Dropping rows with missing values:

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna(axis=0)
cleaned_data_count = recent_grads.shape[0]

print('RAW Counts:' + str(raw_data_count))
print('Cleaned Counts:' + str(cleaned_data_count))
```

# 2 Using a scatter plot to examine some rows:

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample Size Vs Median')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title ='Sample Size vs Unemployment Rate')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full Time vs Median')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Percentage of Women vs Unemployment Rate')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Number of Men Vs Median')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Number of Women vs Median')
```

```python
#one extra plot to explore a relationship between the share of women and median income:
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Percentage of Women vs Median Income')
```

* We can see that the higher the sample size, the lower the median salary in genera. With the majority of high income earners being in majors with the smallest sample sizes.

* Students in majors with majority female employees seem to make less than students in majors with majority male employees.

* There seems to be a slight negative relationship between the number of full time employees in a certain field and the median salary.
